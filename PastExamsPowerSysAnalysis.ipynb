```python
from cmath import pi
from cmath import sqrt
import math
from re import X
from tkinter import Y
import numpy as np
import sympy as smp
from sympy.solvers import solve
from sympy import Symbol
from scipy.optimize import fsolve
from math import exp
from sympy import *
from scipy.optimize import fsolve
from math import exp
import scipy.optimize
import cmath
import operator
def polarTOcart(r,phi):
    x=r*cos(phi*0.0174532925).n()
    y=r*sin(phi*0.0174532925).n()
    return(x,y)
def polar2cart(r,phi):
    x=r*cos(phi*0.0174532925).n()
    y=r*sin(phi*0.0174532925).n()
    return(y,x)    
```

```python
print('How many Loads we have')
n=int(input())
Qtot=0
Ptot=0
for i in range (n):
    print("Give complex power, S of the Load",i+1,'in kVA')
    L1=float(input())
    print("Give pf of Load",i+1)
    pf1=float(input())
    P1=pf1*L1
    print('The total Real Power P',i+1,'=',P1,'kW')
    Q1=smp.sqrt(L1**2-P1**2)
    print('The apparent Power Q',i+1,'=',Q1,'kVAR')
    i=i+1
    Qtot=Qtot+Q1
    Ptot=Ptot+P1
        
Stot=smp.sqrt(Qtot**2+Ptot**2)
pftot=Ptot/Stot
print('The total RealPower Ptot=',round(Ptot,1),'kW')
print('The total ApparentPower Qtot=',round(Qtot,1),'kVAR')
print('The total ComplexPower Stot=',round(Stot,1),'KVA')
print('The total PowerFactor pfTOT=',round(pftot,2),'lagging')
```
